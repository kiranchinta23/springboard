#### Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members!

```python
query1 =("""
SELECT *
FROM (
    SELECT name AS facility,
        SUM(CASE WHEN memid = 0 THEN guestcost * slots
            ELSE membercost * slots END) AS revenues
    FROM Bookings
    INNER JOIN Facilities
        USING (facid)
    GROUP BY facid
    ORDER BY revenues
)
WHERE revenues < 1000
""")


df1=pd.read_sql_query(query,engine)
```

/Q12: Find the facilities with their usage by member, but not guests/

```python
query3=("""
select firstname||','||surname, name as facility,sum(slots*0.5) as usage_hours
from Bookings b
inner join Facilities f
    using (facid)
inner join Members m
    using (memid)
where memid!=0
group by memid,facility
order by facility
""")


df3=pd.read_sql_query(query3,engine)
df3
```

Q13: Find the facilities usage by month, but not guests /

```python
query4=("""
select 
    name as facility,
    strftime('%m',DATE(starttime)) as month,
    sum(slots*0.5) as usage_hours
from Bookings b
inner join Facilities f
    using (facid)
where memid!=0
group by facility,month
order by facility
""")


df4=pd.read_sql_query(query4,engine)
df4
```
